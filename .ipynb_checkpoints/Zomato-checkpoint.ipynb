```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country
```

```python
final_df = pd.merge(df , df_country , on='Country Code' , how = 'left')
```

```python
final_df
```

```python
final_df.Country.value_counts()   ##it means that the zomato probably most transaction happening in india 
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_values = final_df.Country.value_counts().values
```

```python
country_values
```

```python
##Top 3 country uses zomato
plt.pie(country_values[:3] , labels=country_names[:3] , autopct="%1.1f%%" )
```

Zomato Maximum records are from India then United kingdom and after that United state 

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating count'})
```

```python
final_df.columns
```

```python
ratings
```

## OBSERVATION

1.) When rating are between 4.5 to 4.9 --> Excellent
2.) When rating are between 4.0 to 4.4 --> very good
3.) When rating are between 3.5 to 3.9 --> good
4.) When rating are between 2.5 to 3.4 --> Average
5.) When rating are between 1.8 to 2.4 --> Poor

and 0 rating has been given by most of the people

```python
import seaborn as sns
plt.figure(figsize=(12,6))
sns.barplot(x = 'Aggregate rating' , y = 'Rating count' , data = ratings , palette='viridis',saturation=1.5)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x = 'Aggregate rating' , y = 'Rating count' , data = ratings ,hue = 'Rating color', palette=['white' ,'red', 'orange', 'yellow' ,'green','green' ],saturation=1.5)
```

Observation :
 1.) not rated are count is very high
 2.) maximum number of rating between 2.5 to 3.4

```python
sns.countplot(x = 'Rating color' , data = ratings, palette='viridis')
plt.xticks(rotation = 45 )
plt.tick_params(direction='out', length=10, width=2, colors='red')

```
